```python
import pdfplumber
import chromadb

def extract_text_from_pdf(pdf_path):
    """Extracts text and metadata from a single PDF."""
    chunks = []
    metadata = []
    with pdfplumber.open(pdf_path) as pdf:
        for page_num, page in enumerate(pdf.pages, start=1):
            text = page.extract_text()
            if text:
                chunks.append(text)
                metadata.append({"page": f"Page {page_num}", "doc_id": f"{pdf_path}_{page_num}"})
    print(metadata)
    return chunks, metadata

def upsert_into_chromadb(chunks, metadata):
    """Upserts documents, metadata, and embeddings into ChromaDB."""
    collection.add(
        documents=chunks,
        metadatas=metadata,
        ids=[meta['doc_id'] for meta in metadata]
    )

# Initialize ChromaDB client and collection
chroma_client = chromadb.PersistentClient(path="./chromadb")
collection = chroma_client.get_or_create_collection(name="doc_collection")

# List of PDF file paths
pdf_files = ['D:\GenAI\RAG\FSB Bulletin on AI and ML.pdf','D:\GenAI\RAG\OCC - MRM.pdf','D:\GenAI\RAG\SR - 11-7 MRM.pdf']

# Process each PDF and store data in ChromaDB
for pdf_file in pdf_files:
    print(f"Processing {pdf_file}...")
    chunks, metadata = extract_text_from_pdf(pdf_file)
    upsert_into_chromadb(chunks, metadata)

print("Data from all PDFs successfully stored in ChromaDB.")
```

```python
# Initialize ChromaDB client and collection : make sure it's the right collection !
chroma_client = chromadb.PersistentClient(path="./chromadb")
collection = chroma_client.get_or_create_collection(name="doc_collection")

# Step 1: Function to query ChromaDB with a prompt
def query_chromadb(prompt, n_results=3):
    results = collection.query(
        query_texts=[prompt],  # User's query
        n_results=n_results,   # Number of relevant chunks to retrieve
        include=["documents", "metadatas"]  # Retrieve both document text and metadata
    )
    return results

# Step 2: Flattening functions for retrieved documents and metadata
def flatten_documents(documents):
    return [sentence for doc in documents for sentence in doc]

def flatten_metadatas(metadatas):
    return [meta for meta_list in metadatas for meta in meta_list]

```

```python
# Example query and flattening
prompt = "What does the examination process for the 'quantity of risk' cover, and what are the key steps involved?"  # User's question
# prompt = "What is'quantity of risk' cover? what is the examination process and key steps involved"
chromadb_results = query_chromadb(prompt)
flat_chunks = flatten_documents(chromadb_results["documents"])
flat_metadata = flatten_metadatas(chromadb_results["metadatas"])

# Join the retrieved chunks with delimiters and metadata
retrieved_chunks = [f"{chunk} (Source: {meta['page']})" for chunk, meta in zip(flat_chunks, flat_metadata)]
full_retrieved_chunks = "\n\n---\n\n".join(retrieved_chunks)

```
